##### 실습문제


```python
def cal(n1, n2=1, op='+') : # 매개변수 기본값 설정하기
    """덧셈과 뺄셈을 계산하는 함수""" # docstring 추가하기
    if op == '+' :
        return n1+n2
    elif op == '-' :
        return n1-n2
    else :
        return '연산기호 입력 에러'
```

```python
num1 = int(input('첫 번째 정수 입력 >> '))
num2 = int(input('두 번째 정수 입력 >> '))
op = input('연산자 입력(+,-) >> ')
print(f'결과 : {cal(num1, num2, op)}')
```

```python
print(cal(7,op='-'))
print(cal(8))
print(cal('A','B'))
```

```python
def add(num, *args) :
    total = 0
    for n in args :
        total+= n
    return total
```

```python
print(add(1,2,3))
print(add(4,5,6,7))
```
