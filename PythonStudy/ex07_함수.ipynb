```python
def cal(n1, n2=1, op='+') : #   
    """   """ # docstring 
    if op == '+' :
        return n1+n2
    elif op == '-' :
        return n1-n2
    else :
        return '  '
```

```python
num1 = int(input('    >> '))
num2 = int(input('    >> '))
op = input(' (+,-) >> ')
print(f' : {cal(num1, num2, op)}')
```

```python
print(cal(7,op='-'))
print(cal(8))
print(cal('A','B'))
```

```python
def add(num, *args) :
    total = 0
    for n in args :
        total+= n
    return total
```

```python
print(add(1,2,3))
print(add(4,5,6,7))
```
