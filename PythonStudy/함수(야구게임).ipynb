```python
import random
```

```python
def isDuplicate(li) :
    for i in range(len(li)) :
        for k in range(i+1, len(li), 1) :
            if li[i] == li[k] :
                return True
    return False    
```

```python
li1 = [2,7,3,5,8]
check = isDuplicate(li1)
print(check)
```

```python
def createRandomNumber(cnt):
    randomList = [0,0,0]
    while isDuplicate(randomList):
        randomList = []
        for i in range(cnt) :
            randomList.append(random.randint(1,50))

    return randomList
```

```python
randomLi = createRandomNumber(5)
print(randomLi)
```

```python
def strikeBallCounting(li1, li2) :
    strike = 0
    ball = 0
    for i in range(len(li1)) :
        for k in range(len(li1)) :
            if li1[i] == li2[k] :
                if i == k :
                    strike +=1
                else :
                    ball +=1
    return strike, ball
```

```python
li1 = [5,7,8]
li2 = [5,8,7]
strike, ball = strikeBallCounting(li1, li2)
print(strike, ball)
```

```python
print("==GameStart==")
randomLi = createRandomNumber(3)
# print(randomLi)

while True :
    inputLi = []
    for i in range(len(randomLi)) :
        num = int(input(f'{i+1}번째 숫자 입력 : '))
        inputLi.append(num)
    strike, ball = strikeBallCounting(randomLi, inputLi)
    print(f"strike : {strike}\t ball : {ball}\n")
    
    if strike == 3 :
        print('게임종료!!')
        break
```
