- while문


##### 실습문제


```python
while True :
    n1 = int(input('첫 번째 정수 입력 >> '))
    n2 = int(input('두 번째 정수 입력 >> '))
    
    if n1==0 and n2 == 0 :
        print('프로그램 종료')
        break
    print(f'두 수의 합 : {n1+n2}')
```

```python
num1, num2 = 0, 1
while num1 != 0 or num2!=0 :
    num1 = int(input('첫 번째 정수 : '))
    num2 = int(input('두 번째 정수 : '))
    
    if num1 != 0 or num2!=0 :
        print(f"두 수의 합 : {num1+num2}")
print('프로그램 종료')
```
